```python
variables = [r1,r2,r3,r4,r5,r6,r7,r8,r9,r10,r11,r12,r13,r14,r15,r16,r17,r18,r19,r20,r21,r22,r23,r24,r25,r26,r27,r28,r29,r30,r31,r32,r33,r34,r35,r36,r37,r38,r39,r40,r41,r42,r43,r44,r45,r46,r47,r48,r49,r50]
h = 0
bkname = []
bkurl = []
bkprice = []
bkrating = []
while h < 50:
    r = variables[h]
    print(f'r{h}')
    variables[h] = bs(variables[h].content)#Converts to BeautifulSoup
    #Body section
    first_dir = variables[h].body

    #Div section(container_fluid page)
    second_dir = first_dir.find_all("div")

    #Div class="page_inner"
    third_dir = second_dir[3].find("div")

    fourth_dir = third_dir.find_all("div")
    fourth_dir[3]#Div class="col-sm-8 col-md-9"

    #section directory
    fifth_dir = fourth_dir[3].find("section")

    sixth_dir = fifth_dir.find_all("div")#div within the section

    #div container containing the books
    seventh_dir = sixth_dir[1]

    #returns a list of books in the webpage
    book_dir = seventh_dir.find_all("li")
    altered_book_dir = book_dir[0:20]
    h = h+1
    

    i = 0

    while i < 20:
        book = altered_book_dir[i]
        bookNameAndImageUrl = book.find_all("a")
        bookName = bookNameAndImageUrl[1]["title"]#Name of Book
        bkname.append(bookName)
        print(bookName)

        ImageUrl = bookNameAndImageUrl[0]["href"]#Image URL of the book
        bkurl.append(ImageUrl)
        print(ImageUrl)

        priceAndRating = book.select("p")
        x = priceAndRating[0]["class"]
        rating = f"{x[1]} {x[0]}"#Rating of the book
        bkrating.append(rating)
        print(rating)

        book_price = priceAndRating[1].string#Price of the book
        bkprice.append(book_price)
        print(book_price)
        i = i+1
```

```python
#save to pandas dataframe
dict = {"Book Name":bkname, "Image URL":bkurl, "Price":bkprice, "Rating":bkrating}
df = pd.DataFrame(dict)
df.to_csv('webscraping.csv')
```
